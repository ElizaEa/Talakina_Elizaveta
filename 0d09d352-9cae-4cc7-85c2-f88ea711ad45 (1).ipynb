```python
def check(row):
    col = row.columns
    dupli = row.duplicated().sum()
    print('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —è–≤–Ω—ã—Ö –¥—É–±–ª–∏–∫–∞—Ç–æ–≤:', dupli)
    nans = row.isna().sum().sum()    
    print('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ –¥–∞–Ω–Ω—ã—Ö:', nans)
    if dupli > 0:
        print(row.duplicated())
    for col_l in col:
        print('-'* 30)
        print('–£–Ω–∏–∫–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è:', col_l, row[col_l].sort_values().unique())
```

```python
# –í—ã–≤–æ–¥ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º, pie-–¥–∏–∞–≥—Ä–∞–º–º, —Å–≤–æ–¥–Ω—ã—Ö —Ç–∞–±–ª–∏—Ü
def sign(row,column, title):
    if row.dtypes[column] == object:
        row[column].value_counts().plot.barh(figsize=(10, 5), grid= True)
        plt.title(title, fontsize=15)
        plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')
        plt.rc('xtick', labelsize= 15 ) 
        plt.rc('ytick', labelsize= 15 ) 
        plt.show()
        one = pd.pivot_table(
            row,
            index=column,
            values = 'customerID',
            aggfunc='count')
        plt.title(title, fontsize=20)
        row[column].value_counts().plot(
            kind = 'pie',
            figsize=(10,10),
            autopct = '%0.01f%%',
            colors=sns.color_palette('Set2')
        )
        ax = plt.gca()
        ax.axes.xaxis.set_visible(False)
        ax.axes.yaxis.set_visible(False)
        plt.show()
        one['% '] = round(one['customerID']/one['customerID'].sum()*100,2)
        one.rename(columns={'customerID':'–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ'},
              inplace=True)
        print(one)
        #return one

    else:
        fig, axes = plt.subplots(2, figsize=(15, 15), sharey='row')
        #axes[0].plot(row[column], kind='bar')
        axes[0].hist(row[column], bins=int(len(row) ** (0.5)))
        axes[0].grid(True)
        axes[0].set_title(title, fontsize=20)
        axes[0].set_ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')
        axes[1].boxplot(data=row, x=column, vert=False, patch_artist=True)
        axes[1].grid(True)
        axes[1].set_xlabel(title)
        plt.show()
        one = row[column].describe().T
        print(one)
        #return one
```

```python
def pie(row, column ):
    row[column].value_counts().plot.barh(figsize=(10, 5), grid= True)
    plt.title(column, fontsize=15)
    plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')
    plt.rc('xtick', labelsize= 15 ) 
    plt.rc('ytick', labelsize= 15 ) 
    plt.show()
    one = pd.pivot_table(
        row,
        index=column,
        values = 'customerID',
        aggfunc='count')
    plt.title(column, fontsize=20)
    row[column].value_counts().plot(
        kind = 'pie',
        figsize=(10,10),
        autopct = '%0.01f%%',
        colors=sns.color_palette('Set2')
    )
    ax = plt.gca()
    ax.axes.xaxis.set_visible(False)
    ax.axes.yaxis.set_visible(False)
    plt.show()
    one['% '] = round(one['customerID']/one['customerID'].sum()*100,2)
    one.rename(columns={'customerID':'–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ'},
               inplace=True)
    return one
```

```python
def phikmatrix(row, col):
    if col in row.columns:
        phik_overview = (row.drop(col, axis=1).phik_matrix(verbose=False))
    else:
        phik_overview = (row.phik_matrix(verbose=False))
    matrix = np.triu(phik_overview .corr())
    plot_correlation_matrix(
    phik_overview.values,
    x_labels=phik_overview.columns,
    y_labels=phik_overview.index,
    vmin=0, vmax=1, color_map='RdPu',
    title=r'–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è $\phi_K$',
    fontsize_factor=1.5,
    figsize=(20, 15)
) 
```

```python
num_col = []
cat_col = []
def types(row, col, id):
    columns = row.columns
    for column in columns:
        if column != col and column != id:
            if row.dtypes[column] == object:
                cat_col.append(column)
            else:
                num_col.append(column)
    return num_col, cat_col
```

1. –£–±—Ä–∞—Ç—å –∑–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –ø—Ä–æ–ø—É—Å–∫–æ–≤, —Ç.–∫. –≤ –Ω–∞—à–µ–º –¥–∞—Ç–∞—Å–µ—Ç–µ –ø—Ä–æ–ø—É—Å–∫–∏ —É–∂–µ –∑–∞–ø–æ–ª–Ω–µ–Ω—ã. –ö—Ä–æ–º–µ —Ç–æ–≥–æ, –∑–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –Ω–∞–∏–±–æ–ª–µ–µ —á–∞—Å—Ç—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º –≤ –Ω–∞—à–µ–º –∫–æ–Ω–∫—Ä–µ—Ç–Ω–æ–º —Å–ª—É—á–∞–µ —è–≤–ª—è–µ—Ç—Å—è –Ω–µ–ø—Ä–∞–≤–∏–ª—å–Ω—ã–º.
    
2. –ü–æ—Ä—è–¥–∫–æ–≤–æ–µ –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ –¥–ª—è –ª–∏–Ω–µ–π–Ω—ã—Ö –º–æ–¥–µ–ª–µ–π –Ω–µ–ø—Ä–∞–≤–∏–ª—å–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å, –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å OHE –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ.

```python
ohe_pipe = Pipeline(
    [
        (
            'ohe', 
            OneHotEncoder(sparse_output=False,drop='first')
        )
    ]
)

# —Å–æ–∑–¥–∞—ë–º –æ–±—â–∏–π –ø–∞–π–ø–ª–∞–π–Ω –¥–ª—è –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∏ –¥–∞–Ω–Ω—ã—Ö
data_preprocessor = ColumnTransformer(
    [('ohe', ohe_pipe, cat_col),
     ('num', MinMaxScaler(), num_col)
    ],  
    remainder='passthrough'
)

pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models',  DecisionTreeClassifier(random_state=RANDOM_STATE))
     ]
    )
```
